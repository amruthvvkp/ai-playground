## Create a scatter plot

We will use the data in Appendix A to create a scatter plot.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array(dataset, dtype=np.float32)
x = data[:, 0]
y = data[:, 1]

plt.scatter(x, y)
plt.xlabel(headers[0])
plt.ylabel(headers[1])
plt.title("Test 1 vs Test 2")
plt.show()
```

## Fit a best fit line

We will use the data in Appendix A to fit a best fit line.

```python
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(x, y)

def calculate_y(x):
    """Return the y value for a given x value."""
    return slope * x + intercept

mymodel = list(map(calculate_y, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.xlabel(headers[0])
plt.ylabel(headers[1])
plt.title("Test 1 vs Test 2")
plt.show()
```

## Discover the equation for the line

We will use the data in Appendix A to discover the equation for the line.

```python
print(f"y = {slope}x + {intercept}")
print(f"r = {r}")
print(f"p = {p}")
print(f"std_err = {std_err}")
```

## Check the regression and see if it is statistically significant

We will use the data in Appendix A to check the regression and see if it is statistically significant.

```python
if p < 0.05:
    print("The relationship is statistically significant")
else:
    print("The relationship is not statistically significant")
```

## Predict what some one who score 60 in Test 1 will get in Test 2

We will use the data in Appendix A to predict what some one who score 60 in Test 1 will get in Test 2.

```python
print(f"{round(calculate_y(60),2)}% is the predicted score for Test 2 if the score for Test 1 is 60")
```
